# Faster R-CNN 代码介绍与任务总结

# 1.数据集

## 1.1 数目
- 3000个训练样本
- 1000个测试样本
- 且保证两者任何一个数字无相同

## 1.2 格式
- 图片是大小均为 200 x 200 的.jpg文件
- 标签是包含图片地址，以及各个 box 的类别和坐标的.txt文件
- 并且训练集和测试集已经分开

## 1.3 数据集生成规律
- 原数字（28 x 28）
- 尺度等比例放大[0.5,2]倍
- 旋转[-15,15]度
- 所有数字（经以上变换后）左上角坐标间距不小于 36 像素
- 取 3000 中 2750 生成 3000 个训练样本（200 x 200）
- 取剩余 250 生成 1000 个测试样本 （200 x 200）
- 同时精确计算样本的真实 boxes 信息并保存

### 2.1.2 导入以及预处理

```python
def get_data(input_path):
```

输入txt文件(annotation file)地址，返回 3 个参数

例如`train_imgs, classes_count, class_mapping = get_data(train_path)`

- `train_imgs`:列表文件，包含(图片路径, 宽, 高, boxes坐标)

- `classes_count`:字典文件，类别id和数目，{key:class_name, value:count_num} ，e.g. {'Car': 2383, 'Mobile phone': 1108, 'Person': 3745}

- `class_mapping`: 字典文件，类别映射，{key:class_name, value: idx}，e.g. {'Car': 0, 'Mobile phone': 1, 'Person': 2}

```python
def get_anchor_gt(all_img_data, C, img_length_calc_function, mode='train'):
```

输入 `get_data()`的第一个输出文件和一些设置，返回一个生成器

例如`data_gen_train = get_anchor_gt(train_imgs, C, get_img_output_length, mode='train')`

`X, Y, image_data, debug_img, debug_num_pos = next(data_gen_train)`

- `X` : 预处理(包含图像增强)后的 min-batch 默认 `shape=(1, 200, 200, 3)`

- `Y` : 即`[y_rpn_cls, y_rpn_regr]` ,是根据 ground truth 生成的 正负样本

    `y_rpn_cls.shape=(1, 25, 25, 18)`,每一个特征图的一个像素，对应 9 个 anchors, 但是注意到第四维是 18 ，依源代码，在这 18 维向量里，第 n 个和第 n+9 个元素对应一个 anchor ，靠前的元素为 1 表示为前景，0 表示为后景；靠后的元素为 1 表示正样本，0 表示负样本。并且在这一步中，随机选 128 个正负样本（原论文256），具体操作是随机把表示前景的元素由 1 设置为 0 。
    
    `y_rpn_regr.shape=(1, 25, 25, 72)`,回归的坐标，一样的，在 72 维的向量里，第 n 个和第 n+36 个元素对应一个 anchor 的坐标，每 4 个相邻元素就是一个 anchor 的坐标回归值(x1,y1,x2,y2), 若这个 anchor 为后景则默认为 0 。
    
- `image_data`就是每一个min-batch的ground truth, `debug_img`,`debug_num_pos`都是用于 debug 的图片和其正样本数目。

```python
def calc_rpn(C, img_data, width, height, resized_width, resized_height, img_length_calc_function):
```

计算出 anchors 的函数(*重要*)

放在 `get_anchor_gt()`函数中使用

`y_rpn_cls, y_rpn_regr, num_pos = calc_rpn(C, img_data, width, height, resized_width, resized_height, img_length_calc_function)`

- 输入中`img_data`包含各种图片信息(同`get_anchor_gt()`的第一个输入，或者使用数据增强函数`augment()`后的输出)

- `width, height, resized_width, resized_height`,分别是原图像对应宽高和预处理后的宽高

- `img_length_calc_function`是计算出特征图维度的函数，例如输入`(200,200,3)`，输出`(25,25,128)`

处理时，先循环所有所有 anchors (以 25 x 25 x 9 方式先后迭代)，然后对每一个 anchor 计算和所有 ground truth boxes 的 iou，默认所有 anchors 为负样本，取`curr_iou > C.rpn_max_overlap`的为正样本，`C.rpn_min_overlap < curr_iou < C.rpn_max_overlap`的为后景，并且确保所有ground truth boxes 至少对应一个正样本。另外只有正样本才计算回归值(tx,ty,tw,th)，否则默认为 0.

返回的就是[y_rpn_cls,y_rpn_regr]

NMS算法

输入 boxes 及其概率分布

返回 max_boxes 个重叠度较小的ROIs

算法步骤大致就是
1. 找出概率最大的 box 并 append 进输出
2. 删除此 box 及其重叠面积大于某阈值的 boxes
3. 循环步骤1和2直到列表为[]

### 2.2.2 训练模型详细步骤

训练设置：

- total_epochs = 23
- epoch_length = 1000
- Adam(lr=0.0001)

`for epoch_num in range(num_epochs):`

`X, Y, img_data, debug_img, debug_num_pos = next(data_gen_train)` ：获取数据(`X.shape=(1,200,200,3)`, `Y=[y_rpn_cls, y_rpn_regr]`)
    
`loss_rpn = model_rpn.train_on_batch(X, Y)` ：训练 RPN 网络,`loss_rpn`即`[_, loss_rpn_cls, loss_rpn_regr]`
        
`P_rpn = model_rpn.predict_on_batch(X)` ：获取 RPN 网络的预测结果 `[rpn_cls, rpn_regr]`

`R = rpn_to_roi(P_rpn[0], P_rpn[1], C, K.image_dim_ordering(), use_regr=True, overlap_thresh=0.7, max_boxes=300)` ：
将 RPN 预测的结果转换为 ROIs ，利用 NMS 算法提取了 300 个,`R.shape=(300,4)`

在与 ground truth boxes 的 IoUs 大于某阈值（0.1）的 ROIs 中随机选取`num_rois=2`个ROIs(其中一半为正，一半为负样本)，之后训练分类网络

`loss_class = model_classifier.train_on_batch([X, X2[:, sel_samples, :]], [Y1[:, sel_samples, :], Y2[:, sel_samples, :]])`：输入 `X` 为原图像(1,200,200,3)，`X2[:, sel_samples, :]` 为随机的正负样本(1,2,4), `Y1,Y2`为对应 ground truth 的类别和坐标(1,2,11),(1,2,80),注意 Y1 包括后景的分类，而 Y2 的数据结构和之前提到的类似，80 维向量里，第 n 和第 n+40 元素对应一个坐标，相邻 4 个元素组成一个 box。前 40 元素为 1 表示其不是后景，默认为 0，后 40 个元素就对应每个类的 box。注意到，`model_classifier`的输入是 img 而不是特征图，也就是说共享卷积层在一个 min-batch 中被训练了两次，分别包含 PRN 和 classifier 的反向传播信息。
