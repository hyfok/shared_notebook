```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
plt.style.use('ggplot')
```

```python
x_f = np.array([0.5,1.0,2.0,3.0,5,7,9,11,13,15,18,22,26,30],dtype=np.float32)
y_ur = np.array([1.79,1.78,1.75,1.71,1.59,1.49,1.41,1.27,1.25,1.19,1.12,0.985,0.872,0.776],dtype=np.float32)
```

```python
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=15)
```

```python
plt.figure(figsize=(10,6))
plt.title(u"幅频特性曲线", fontproperties=font)

x_new = np.linspace(x_f.min(),x_f.max(),300)
func = interpolate.interp1d(x_f,y_ur,kind='cubic')
y_new = func(x_new)

#plt.plot(x_f, y_ur, 'o-', linewidth=3)
plt.plot(x_new, y_new, linewidth=3)
plt.plot(x_f,y_ur,'o')
plt.xlabel('f/kHz')
plt.ylabel('Uc/v')

plt.legend()
```

```python
y_tan = np.array([0.1,0.138,0.182,0.250,0.477,0.726,0.781,1.06,1.16,1.32,1.54,1.64,1.75,2.12],dtype=np.float32)
```

```python
from sklearn.linear_model import LinearRegression
cls = LinearRegression()
cls.fit(x_f.reshape(-1, 1),y_tan.reshape(-1, 1))
#cls.coef_,cls.intercept_

plt.figure(figsize=(10,6))
plt.title(u"RC串联电路tan"+r'$\Phi$-f', fontproperties=font)

x_new = np.linspace(x_f.min(),x_f.max(),300)
func = interpolate.interp1d(x_f,y_tan,kind='cubic')
y_new = func(x_new)

#plt.plot(x_f, y_ur, 'o-', linewidth=3)
plt.plot(x_new, y_new, linewidth=3)
plt.plot(x_f,y_tan,'o')
plt.xlabel('f/kHz')
plt.ylabel( r'tan$\Phi$')

y_a = x_f*float(cls.coef_)+float(cls.intercept_)
plt.plot(x_f,y_a)
r_2 = np.sum((y_tan-y_tan.mean())*(y_a-y_a.mean()))/np.sqrt(np.sum((y_tan-y_tan.mean())**2)*np.sum((y_a-y_a.mean())**2))
plt.text(2,2,"y = %.4fx + %.4f\nR^2 = %.4f"%(cls.coef_,cls.intercept_,r_2))

plt.legend()
```
