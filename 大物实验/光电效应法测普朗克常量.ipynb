```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=15)
plt.style.use('ggplot')
```

```python
y_ua = np.array([1.901, 1.407, 1.145, 0.666, 0.583])
x_v = np.array([8.214, 7.408, 6.879, 5.490, 5.196])*10**14
```

```python
from sklearn.linear_model import LinearRegression
cls = LinearRegression()
cls.fit(x_v.reshape(-1, 1),y_ua.reshape(-1, 1))

plt.figure(figsize=(10,6))
plt.title(u"Uo ~ v 曲线", fontproperties=font)

plt.plot(x_v, y_ua, 'o')
y_a = np.linspace(0,9.2e14,5)*float(cls.coef_)+float(cls.intercept_)
plt.plot(np.linspace(0,9.2e14,5),y_a)
#r_2 = np.sum((y_tan-y_tan.mean())*(y_a-y_a.mean()))/np.sqrt(np.sum((y_tan-y_tan.mean())**2)*np.sum((y_a-y_a.mean())**2))
```
