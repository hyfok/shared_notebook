```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img_bgr = cv2.imread('people.png')
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img_bgr)
plt.subplot(122)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
```

```python
plt.imshow(corners, cmap='gray')
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp = sift.detect(img_bgr)
```

```python
import numpy as np
img_kp = np.zeros_like(img_bgr)
img_kp = cv2.drawKeypoints(img_bgr, kp, img_kp, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_kp)
```

```python
kp, des = sift.compute(img_bgr, kp)
```

```python
des.shape
```

```python
kp2, des2 = sift.detectAndCompute(img_bgr, None)
```

```python
np.allclose(des,des2)
```

```python
surf = cv2.xfeatures2d.SURF_create()
kp, des = surf.detectAndCompute(img_bgr, None)
img_kp = cv2.drawKeypoints(img_bgr, kp, img_kp, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_kp)
```
