```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#导入相应库
```

```python
img=cv2.imread('lena.jpg')
for i in range(9): #最多迭代9次，即生成9层的金字塔
    img=cv2.GaussianBlur(img,(5,5),15) #利用opencv的高斯滤波对原图像卷积（核5*5），产生高斯平滑
    img = img[0:-1:2,0:-1:2] #对图像向下采样，约缩小为上一幅的1/2
    plt.subplot(3, 3, i + 1) #以下均为显示相关操作
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
    plt.axis('off')
```
